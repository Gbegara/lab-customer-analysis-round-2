```python
import pandas as pd
import numpy as np
```

```python
mca = pd.read_csv('marketing_customer_analysis.csv')
display(mca.head())
mca.shape
```

```python
mca = mca.drop(['Unnamed: 0'], axis = 1)
```

```python
display(mca.head())
```

```python
mca = mca.drop_duplicates()
mca.shape
```

```python
cols1 = []
for column in mca.columns:
    cols1.append(column.lower())
mca.columns = cols1
display(mca.head())
```

```python
mca._get_numeric_data()
```

```python
mca.select_dtypes(include= 'object').columns
```

```python
mca.isna().sum()
```

```python
mca[mca['state'].isna()==True].head(60)
```

```python
#I asumme that values NaN are 'No' responses, the same with claims, and open complains, putting those two as 0
noresponse= 'No'
mca['response'] = mca['response'].fillna(noresponse)
mca.isna().sum()
noclaims= 0
mca['months since last claim'] = mca['months since last claim'].fillna(noclaims)
mca.isna().sum()
nocomplaints= 0
mca['number of open complaints'] = mca['number of open complaints'].fillna(nocomplaints)
mca.isna().sum()
```

```python
#Checking number of vehicle class, size and type to see which are the most commons
mca['vehicle size'].unique()
mca['vehicle size'].value_counts()
```

```python
#I was going to erase the column but vehicle class 'A' can be a good information for those who have it
not_a='Not A'
mca['vehicle type'] = mca['vehicle type'].fillna(not_a)
mca
```

```python
#assigning the most common values to the NaN
vhclass= 'Four-Door Car'
mca['vehicle class'] = mca['vehicle class'].fillna(vhclass)
vhsize= 'Medsize'
mca['vehicle size'] = mca['vehicle size'].fillna(vhsize)

mca.isna().sum()
```

```python
mca['state'].unique()
mca['state'].value_counts()
```

```python
#Since state has many of them I think that data could be usefull, we are not going to be able to compare anything with state
#but other data can be usefull
unstate= 'Unknown'
mca['state'] = mca['state'].fillna(unstate)
mca.isna().sum()
```

```python
#changing effective to datetime type
from datetime import date, timedelta

mca['effective to date'] = pd.to_datetime(mca['effective to date'], errors='coerce')
mca
```

```python
#found this function because with the one in the class didnt work "https://stackoverflow.com/questions/26105804/extract-month-from-date-in-python"
mca['month'] = mca['effective to date'].dt.month
mca.dtypes

```

```python
#normalizing the dataframe
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
transformer = MinMaxScaler().fit(X)
X_normalized = transformer.transform(X)
X_normalized= pd.DataFrame(X_normalized, columns=X.columns)
X_normalized


```

```python
#checking different columns to choose which ones to drop
mca_cat['vehicle class'].unique()
mca_cat['vehicle class'].value_counts()
```

```python
#dropping columns that I consider that they influence very little in the result
mca_cat = mca_cat.drop(['vehicle type','education', 'customer', 'marital status', 'policy','employmentstatus', 'response'], axis=1)
mca_cat.head()
```

```python
#dividing between those columns that I am going to do a hot encoded an a ordinal encoded
X_cat_hot = mca_cat.drop(['coverage', 'renew offer type'], axis=1)

X2_cat_ord = mca_cat[['coverage', 'renew offer type']]

display(X_cat_hot.head())
display(X2_cat_ord.head())
```

```python
#hot encoding
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import LabelEncoder

hot_encoder = OneHotEncoder().fit(X_cat_hot)
print(hot_encoder.categories_)
hot_encoded = hot_encoder.transform(X_cat_hot).toarray()
cols = [colname for row in hot_encoder.categories_ for colname in row]
cols
mca_hot_encoded = pd.DataFrame(hot_encoded,columns=cols)
cols_to_drop = [row[0] for row in hot_encoder.categories_]
cols_to_drop
mca_hot_encoded = mca_hot_encoded.drop(cols_to_drop, axis=1)
mca_hot_encoded


```

```python
#ordinal encoding
from sklearn.preprocessing import OrdinalEncoder
ord_encoded = OrdinalEncoder(categories = [['Basic', 'Extended', 'Premium'], ['Offer1', 'Offer2', 'Offer3', 'Offer4']]).fit(X2_cat_ord)
mca_ord_encoded = pd.DataFrame((ord_encoded.transform(X2_cat_ord)), columns = X2_cat_ord.columns)
mca_ord_encoded.head()
```

```python
#concatenating the 3 different dataframes
X_mca= pd.concat([X_normalized, mca_hot_encoded, mca_ord_encoded], axis=1)
display(X_mca.head())
print(X_mca.shape)
```
